```python
user_total=np.zeros(8141)
cnt=np.zeros(8141,dtype='int64')
```

```python
rating=new_df
```

```python
for i,j in rating.iterrows():
    k=int(j[0])
    cnt[k]+=1
```

```python
import math
test=[math.floor(0.2*i) for i in cnt]
holdout=[math.floor(0.1*i) for i in cnt]
```

```python

n_test=np.array(test,dtype='int64')
n_holdout=np.array(holdout,dtype='int64')
abcd=test+holdout
n_train=[j-i for i,j in zip(abcd,cnt)]
```

```python
tr_temp=pd.DataFrame()
hold_temp=pd.DataFrame()
te_temp=pd.DataFrame()
```

```python
inc_train=np.zeros(8141,dtype='int64')
inc_hld=np.zeros(8141,dtype='int64')
```

```python
for i,j in rating.iterrows():
    user=int(j[0])
    if inc_train[user] < n_train[user]:
        tr_temp=tr_temp.append(j)
        inc_train[user]+=1
    elif inc_hld[user] < n_holdout[user]:
        hold_temp=hold_temp.append(j)
        inc_hld[user]+=1
    else:
        te_temp=te_temp.append(j)
```

```python
te_temp=te_temp.sample(frac=1).reset_index(drop=True)
tr_temp=tr_temp.sample(frac=1).reset_index(drop=True)
hold_temp=hold_temp.sample(frac=1).reset_index(drop=True)

```

```python
movies_mean=np.zeros(shape=[9318,3])
```

```python
user_total=np.zeros(8141)
cnt=np.zeros(8141,dtype='int64')
for i,j in tr_temp.iterrows():
    k=int(j['userId'])
    cnt[k]+=1
    user_total[k]+=j['rating']
    p=int(j['itemId'])
    movies_mean[p][0]+=1
    movies_mean[p][1]+=j['rating']
```

```python
user_mean=np.array([ i/j for i,j in zip(user_total,cnt) ])
for x in movies_mean:
    x[2]=round(x[1]/x[0],2)
    
for x in movies_mean:
    if math.isnan(x[2]):
        x[2]=0
```
